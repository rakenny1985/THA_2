```
from newspaper import Article
odell= Article(urls[1])
odell.download()
odell.parse()

print("Whole data: ", odell)
print("Authors: ", odell.authors)
print("Publish Date: ", odell.publish_date)
print("Article Text: ", odell.text)
```

```
import nltk
nltk.download('punkt')

odell.nlp()
print("Keywords of the article: ", odell.keywords)
print("Article Summary: ", odell.summary)
```

```
details_np = pd.DataFrame(columns=["url", "text", "keywords", "summary"])

for link in urls:
  details = []
  
 
  details.append(link)
  

  odell = Article(link)
  
 
  odell.download()
  
  
  odell.parse()
  
  details.append(odell.text)
  
  odell.nlp()
  
 
  details.append(odell.keywords)
  
 
  details.append(odell.summary)
  
 
  
  details_np.loc[len(details_np)] = details

  
details_np
```
